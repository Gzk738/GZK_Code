```python
import transformers
import datasets
import shap
import numpy as np
```

```python
dataset = datasets.load_dataset("imdb", split="test")

# shorten the strings to fit into the pipeline model
short_data = [v[:500] for v in dataset["text"][:20]]
```

```python
classifier = transformers.pipeline('sentiment-analysis', return_all_scores=True)
classifier(short_data[:2])
```

```python
# define the explainer
explainer = shap.Explainer(classifier)

# explain the predictions of the pipeline on the first two samples
text = ['It was a fantastic performance !', 
        'Best film ever',
        'Such a great show!', 
        'It was a horrible movie',
        'I\'ve never watched something as bad',
        'That is a terrible movie.',
        'us men have right touch in relay duel against australia thens , aug . 17 - so michael phelps is not going to match the seven gold medals won by mark spitz . and it is too early to tell if he will match aleksandr dityatin , the soviet gymnast who won eight total medals in 1980 .']
shap_values = explainer(text)
shap.plots.text(shap_values[:,:,"POSITIVE"])
```
